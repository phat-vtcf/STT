### Convert the Audio File
To save a small headache when using the ```stt``` library, we're going to convert our audio file to (or make sure it is) a 16kHz-sampled *.wav* file.

````python
!ffmpeg -i $AUDIO_FILE_PATH -c:a pcm_s16le -ar 16000 16$AUDIO_FILE_PATH
# ffmpeg won't alter a file in-place, so we'll have to create a new file 
# (namely, 16<file_name>), remove the original, and then rename the new file
!rm $AUDIO_FILE_PATH
!mv 16$AUDIO_FILE_PATH 16$AUDIO_FILE_PATH
````
